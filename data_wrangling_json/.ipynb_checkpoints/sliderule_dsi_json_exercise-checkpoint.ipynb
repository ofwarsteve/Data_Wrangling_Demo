## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties');
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']]);
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# This Cell answers question 1: Find the 10 countries with the most projects #


import matplotlib.pyplot as plt

full_df = pd.read_json('data/world_bank_projects.json')

top_ten = full_df.countryname.value_counts().head(10)
#print(full_df.countryname.value_counts().head(10))

chart = top_ten.plot(kind = 'barh')
chart.set_title('Top Ten Countries by Number of Projects', fontsize = 14)
chart.invert_yaxis()
chart.set_xticks([0, 5, 10, 15, 20])
chart.figsize = (3,4)
chart.position =0.5

for i in chart.patches:
    chart.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)))


```

```python
# This Cell answers question 2: Find the top 10 major project themes #

import numpy as np
import re

full_df = pd.read_json('data/world_bank_projects.json')

# First, I noticed that the major themes column sometimes contains multiple entries. #
# So first, those entries must be unpacked so they can be counted. #

new_df = full_df.mjtheme_namecode.apply(pd.Series)
melted = new_df.melt().dropna()
temp_list = melted.value.tolist()
detail_df = pd.DataFrame.from_dict(temp_list)


# Here, I was inspecting the differences between the melted and detailed df's. #
# It seems that in some cases, the code number is present, but the name is not. # 
# The following code is commented out but can be reviewed to show the missing entries #


#print(detail_df[detail_df.name == ''].count())


# Let's survey the data to see if the code is ever missing #


#print(detail_df[detail_df.code == ''].count())


# Lovely, a code is represented for every row in the dataframe.  So all we need to is identify the top 10 codes, #
# I've done this by isolating the values counts, and selecting the top 10.  Making a temporary df, and then #
# merging this df of top 10 codes with the larger df.  Finally, select only unique code names #

identified_codes = detail_df.code.value_counts()
id_codes = identified_codes.head(10).index.tolist()
id_df = pd.DataFrame(id_codes)
id_df.columns = ['code']

names = pd.merge(id_df, detail_df, on='code').name.unique().tolist()
del(names[1])

print('Top Ten Project Reasons:')
print('    ')
for i in names:
    print(i)
```

```python
import warnings
warnings.filterwarnings('ignore')

# This cell addresses the missing codenames and appropriately filling in the df codenames #
# This should be possible by sorting the df by code, then sorting by codename, then ___ filling the '' entries #

sorted_df = detail_df.sort_values(['code', 'name'])
sorted_df = sorted_df.replace('', np.nan, regex=True).fillna(method='bfill')

#print(sorted_df)


# Sadly, this is pretty far removed from the original dataframe.  This work needs to be applied to the original df #
# Since the new_df is the last place the original index was in force, we'll start there. #

new_df.columns = ['code1', 'code2', 'code3', 'code4', 'code5']
new_df.fillna('Null')

def code_switch(column):
    cseries = column.apply(pd.Series)
    sorted_col = cseries.sort_values(['code', 'name'])
    filled_col = sorted_col.replace('', np.nan, regex = True).fillna(method='bfill')
    return filled_col

col1 = code_switch(new_df.code1)
col2 = code_switch(new_df.code2).drop(0, axis=1)
col3 = code_switch(new_df.code3).drop(0, axis=1)
col4 = code_switch(new_df.code4).drop(0, axis=1)
col5 = code_switch(new_df.code5).drop(0, axis=1)

columns = [col1, col2, col3, col4, col5]

updated_df = col1.merge(col2, how='outer', left_index=True, right_index=True).merge(
    col3, how='outer', left_index=True, right_index=True).merge(
    col4, how='outer', left_index=True, right_index=True).merge(
    col5, how='outer', left_index=True, right_index=True)

updated_df.columns = ['1st_code', '1st_name', '2nd_code', '2nd_name', '3rd_code', '3rd_name', '4th_code', '4th_name', '5th_code', '5th_name']

updated_df
```

```python
# Finally, combine processed codes and namecodes with the original dataframe from the json file.

final_df = full_df.merge(updated_df, left_index=True, right_index=True)
final_df;
```
